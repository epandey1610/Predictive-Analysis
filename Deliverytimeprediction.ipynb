```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
df = pd.read_csv('/Users/eshanakshata/Desktop/delivery_time.csv')
```

```python
df
```

```python
#drop th missing values
df.dropna()
```

```python
# Split the data into X (Sorting Time) and y (Delivery Time)
X = df[['Sorting Time']]
y = df['Delivery Time']
```

```python
# Build a linear regression model
lr_model = LinearRegression()
lr_model.fit(X, y)
y_pred = lr_model.predict(X)
```

```python
# Calculate the RMSE for the initial model
rmse_initial = np.sqrt(mean_squared_error(y, y_pred))
print(f"RMSE (initial model): {rmse_initial}")
```

```python
# Apply logarithmic transformation
X_log = np.log1p(X)
lr_model_log = LinearRegression()
lr_model_log.fit(X_log, y)
y_pred_log = lr_model_log.predict(X_log)
rmse_log = np.sqrt(mean_squared_error(y, y_pred_log))
print(f"RMSE (log transformation): {rmse_log}")
```

```python
# Apply square root transformation
X_sqrt = np.sqrt(X)
lr_model_sqrt = LinearRegression()
lr_model_sqrt.fit(X_sqrt, y)
y_pred_sqrt = lr_model_sqrt.predict(X_sqrt)
rmse_sqrt = np.sqrt(mean_squared_error(y, y_pred_sqrt))
print(f"RMSE (sqrt transformation): {rmse_sqrt}")
```

```python
#Both the logarithmic and square root transformations result in slightly lower RMSE values compared to the initial model. 
#This suggests that these transformations may improve the model's performance in predicting delivery time based on sorting time. 
#However, the improvements are relatively small
```
